```python
import numpy as np, pandas as pd
```

```python
df = pd.read_csv('amazon_reviews.tsv', sep = '\t')
# or else: df = pd.read_table('amazon_reviews.tsv')
df
```

```python
df['label'].value_counts()
```

```python
import nltk
```

```python
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

#### Demonstration:

```python
sia.polarity_scores('Good!!!')
```

## Implementation:

```python
scores = []
for i in range(0,len(df)):
    scores.append(sia.polarity_scores(df['review'][i]))
scores
```

```python
for i in pd.DataFrame(scores):
    df[i] = pd.DataFrame(scores)[i].values
df
```

```python
pred_label = []
for i in df['compound']:
    if i >= 0:
        pred_label.append('pos')
    else:
        pred_label.append('neg')
df['Predicted Label'] = pred_label
df
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(df['label'], df['Predicted Label']))
```

```python
df.to_csv('Sentiment Prediction.csv')
```
